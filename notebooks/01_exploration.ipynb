## 1. Chargement des données et aperçu

````python
# =====================================================================
# Notebook 01 - Exploration des données
# Objectif : prendre connaissance de la table principale et mesurer le
#            déséquilibre de la cible, avant le feature engineering.
# =====================================================================

# Librairies de manipulation de données et de visualisation
import pandas as pd                 # manipulation de tableaux (DataFrames)
import numpy as np                  # calcul numérique
import matplotlib.pyplot as plt     # graphiques de base
import seaborn as sns               # graphiques statistiques (plus jolis)

# Afficher jusqu'à 200 colonnes dans les aperçus (sinon pandas en cache)
pd.set_option("display.max_columns", 200)

# Charger la table principale (1 ligne = 1 client, avec la cible TARGET)
df = pd.read_csv("../data/raw/application_train.csv")

print("Dimensions (lignes, colonnes) :", df.shape)   # attendu : (307511, 122)
df.head()   # aperçu des 5 premières lignes pour visualiser le contenu
````

## 2. Déséquilibre de la cible

C'est **le** point métier central : il y a beaucoup plus de bons payeurs (0) que de mauvais (1).

````python
# --- Déséquilibre de la cible : proportion de bons (0) et mauvais (1) payeurs ---
# C'est LE point métier central : ~92 % de 0 contre ~8 % de 1.
print(df["TARGET"].value_counts(normalize=True))   # proportions de chaque classe

# Visualiser ce déséquilibre sous forme de barres
df["TARGET"].value_counts().plot(kind="bar", title="0 = remboursé, 1 = défaut")
plt.ylabel("Nombre de clients")
plt.show()
````

## 3. Valeurs manquantes

Ces données contiennent beaucoup de valeurs manquantes : on repère les colonnes les plus « trouées ».

````python
# --- Valeurs manquantes : repérer les colonnes les plus "trouées" ---
# isnull().mean() = proportion de NaN par colonne ; *100 pour l'avoir en %.
manquants_pct = df.isnull().mean().sort_values(ascending=False) * 100
print(manquants_pct.head(20))   # les 20 colonnes avec le plus de valeurs manquantes
````

## 4. Piste visuelle : l'âge discrimine le risque

`DAYS_BIRTH` est en jours négatifs (jours écoulés avant la demande). On le convertit en âge positif pour visualiser sa distribution selon la classe.

````python
# --- Piste visuelle : l'âge discrimine bien le risque de défaut ---
# DAYS_BIRTH est en jours NÉGATIFS (jours écoulés avant la demande).
# On le convertit en âge positif (en années) pour le rendre lisible.
df["AGE"] = (-df["DAYS_BIRTH"] / 365).round(1)

# Densité de l'âge selon la classe ; common_norm=False => chaque classe normalisée à part
sns.kdeplot(data=df, x="AGE", hue="TARGET", common_norm=False)
plt.title("Distribution de l'âge selon le remboursement")
plt.show()
````
